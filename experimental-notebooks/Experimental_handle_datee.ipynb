```python
import numpy as np
import pandas as pd
import json
```

```python
def read_data(path):
    '''
    this function will read the txt file which is a line-delimited json file and produce a pandas dataframe.
    
    '''
    my_list =[]
    with open(path) as f:
        for line in f:
            json_content = json.loads(line)
            my_list.append(json_content)
    #df = pd.DataFrame(my_list) 
    return(pd.DataFrame(my_list))
    
```

Handling dates


Different systems stores dates in different formats: 11.12.2019, 2016-02-12, Sep 24, 2003 etc. But for building models on dates data, we need to somehow convert it to a numeric format.
There are 3 most common methods to transform date to numeric format:
Unix timestamp , KSP date format , Divide into several features ,Construct new feature
For this dataset , the third method is used 'Divide into several features ' because :

perfectly preserve intervals
easy intuition and knowledge discovery
more dimensions you add, more complex your model could get, but it is not always bad.


### Convert our date columns to datetime

all the features mentiond in date_col are recognized as objects so in the first place we need to convert them into datetime.
Just beacuse we can tell that these are dates doesn't mean that Python knows that they're dates. Notice that the at the bottom of the output of head(), you can see that it says that the data type of these columns are "object".
Now that we know that our date columns are not being recognized as a date, it's time to convert it so that it is recognized as a date. This is called "parsing dates" because we're taking in a string and identifying its component parts.


```python
Date_col=['accountOpenDate', 'transactionDateTime', 'currentExpDate' , 'dateOfLastAddressChange']
```

```python
# create a new column, date_parsed, with the parsed dates
clean_data['parsed_accountOpenDate'] = pd.to_datetime(clean_data['accountOpenDate'], format = "%Y-%m-%d")
```

```python
# create a new column, 'parsed_currentExpDate', with the parsed dates
clean_data['parsed_currentExpDate'] = pd.to_datetime(clean_data['currentExpDate'], format = "%m/%Y")
```

```python
# create a new column, 'parsed_dateOfLastAddressChange', with the parsed dates
clean_data['parsed_dateOfLastAddressChange'] = pd.to_datetime(clean_data['dateOfLastAddressChange'], format = "%Y-%m-%d")
```

```python
clean_data['parsed_dateOfLastAddressChange'].head()
```

```python
# create a new column, 'parsed_transactionDateTime', with the parsed dates
clean_data['parsed_transactionDateTime'] = pd.to_datetime(clean_data['transactionDateTime'], format = "%Y-%m-%dT%H:%M:%S")
```

```python
clean_data['parsed_transactionDateTime'].head()
```

```python
from datetime import datetime
```

##divide the feature 'parsed_transactionDateTime' into two new features due to the reasons mentioned earlier

```python

clean_data["transactionDateTime_year"] = [t.year for t in clean_data['parsed_transactionDateTime']]
clean_data["transactionDateTime_month"] = [t.month for t in clean_data['parsed_transactionDateTime']]
clean_data["transactionDateTime_day"] = [t.day for t in clean_data['parsed_transactionDateTime']]

clean_data["transactionDateTime_hour"] = [t.hour for t in clean_data['parsed_transactionDateTime']]
clean_data["transactionDateTime_minute"] = [t.minute for t in clean_data['parsed_transactionDateTime']]
clean_data["transactionDateTime_second"] = [t.second for t in clean_data['parsed_transactionDateTime']]

clean_data.head()
```

```python
### adding a feature year_open_date to show the year of opening account(account open date )
```

```python
year_open_date = lambda x: datetime.strptime(x, "%Y-%m-%d" ).year
clean_data['year_open_date'] = clean_data['accountOpenDate'].map(year_open_date)
clean_data.head()
```

```python
## adding a feature month_open_date to show the month of opening account(account open date )
```

```python
month_open_date = lambda x: datetime.strptime(x, "%Y-%m-%d" ).month
clean_data['month_open_date'] = clean_data['accountOpenDate'].map(month_open_date)
clean_data.head()
```

```python
## adding a feature weekday_open_date to show the day of opening account(account open date )
```

```python
weekday_open_date = lambda x: datetime.strptime(x, "%Y-%m-%d" ).strftime('%V')
clean_data['weekday_open_date'] = clean_data['accountOpenDate'].map(weekday_open_date)
clean_data.head()
```

```python
#  'parsed_currentExpDate' and split into three component year month and day
clean_data["currentExpDate_year"] = [t.year for t in clean_data['parsed_currentExpDate']]
clean_data["currentExpDate_month"] = [t.month for t in clean_data['parsed_currentExpDate']]
clean_data["currentExpDate_day"] = [t.day for t in clean_data['parsed_currentExpDate']]
clean_data.head()
```

```python
# work on dateOfLastAddressChange and split into three component year month and day
clean_data["dateOfLastAddressChange_year"] = [t.year for t in clean_data['parsed_dateOfLastAddressChange']]
clean_data["dateOfLastAddressChange_month"] = [t.month for t in clean_data['parsed_dateOfLastAddressChange']]
clean_data["dateOfLastAddressChange_day"] = [t.day for t in clean_data['parsed_dateOfLastAddressChange']]
```
